# Price

```python
df['price'].value_counts()
```

```python
df['price'] = df['price'].str[1:]

def prices(x):
    if x[0] == 'all':
        return np.nan
    
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2) 
        elif x[1] == 'Cr':
            return round(float(x[0]),2)

df['price'] = df['price'].str.split(' ').apply(prices)
# values of prices in crores
```

# Square Feet

```python
temp = df['square_feet'].str.split(" ").str.get(1)
```

```python
temp.value_counts()
```

```python
L = ['sqft','sqyrd','sqm']
df[~temp.isin(L)]
```

```python
df = df[temp.isin(L)]
```

```python
df['square_feet'] = df['square_feet'].str.strip()
df['square_feet'] = df['square_feet'].str.replace(",","")
```

```python
df['square_feet']
```

```python
def areas(x):
    if (x==np.nan) | (x==None):
        return np.nan
    else:
        if type(x) == float:
            return x
        else:
            a = x.split(" ")
            if a[1].strip() == 'sqft':
                return round(float(a[0]),2)
            elif a[1].strip() == 'sqyrd':
                return round(float(a[0])*9,2)
            elif a[1].strip() == 'sqm':
                return round(float(a[0])*10.764,2)
```

```python
df['square_feet'] = df['square_feet'].apply(areas)
```
