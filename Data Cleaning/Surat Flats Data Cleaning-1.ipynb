```python
df['price'] = df['price'].str[1:]

def prices(x):
    if x[0] == 'all':
        return np.nan
    
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2) 
        elif x[1] == 'Cr':
            return round(float(x[0]),2)

df['price'] = df['price'].str.split(' ').apply(prices)
# values of prices in crores
```
