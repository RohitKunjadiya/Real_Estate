```python
new_df = df.merge(ll_df, on='location')

new_df.shape
```

```python
new_df.columns
```

```python
group_df = new_df.groupby('location')[['price','price_per_sqft', 'built_up_area','latitude','longitude']].mean()
```

```python
group_df[['price', 'price_per_sqft']] = group_df[['price', 'price_per_sqft']].round(2)
group_df['built_up_area'] = group_df['built_up_area'].round()
```

```python
import plotly.express as px
import plotly.graph_objects as go
from shapely.geometry import Point
import geopandas as gpd
import pandas as pd

# Helper function to create circles as polygons with a radius of 2.5 km
def create_circle_gdf(df, radius_km):
    gdf = gpd.GeoDataFrame(df, geometry=[Point(xy) for xy in zip(df['longitude'], df['latitude'])], crs="EPSG:4326")
    gdf = gdf.to_crs(epsg=3857)  # Project to meters for buffering
    gdf['geometry'] = gdf['geometry'].buffer(radius_km * 1000)  # Radius in meters
    gdf = gdf.to_crs(epsg=4326)  # Convert back to lat/lon
    return gdf

# Convert the DataFrame to a GeoDataFrame with circles
circle_gdf = create_circle_gdf(group_df, 2)

# Create map with price per square foot as the main layer
fig = px.scatter_mapbox(
    group_df,
    lat='latitude',
    lon='longitude',
    color='price_per_sqft',
    size='built_up_area',
    color_continuous_scale=px.colors.cyclical.IceFire,
    zoom=10,
    size_max=15,
    mapbox_style="open-street-map",
    range_color=[1500, 12000],
    hover_name=group_df.index
)

# Add circle layers by plotting the coordinates of each circle's perimeter
for _, row in circle_gdf.iterrows():
    circle_coords = row['geometry'].exterior.coords.xy
    fig.add_trace(go.Scattermapbox(
        lat=list(circle_coords[1]),
        lon=list(circle_coords[0]),
        mode='lines',
        line=dict(width=1, color='white'),
        fill='toself',
        fillcolor="rgba(255, 0,0, 0.09)",
        showlegend=False
    ))

fig.update_layout(height=600)
fig.show()
```
