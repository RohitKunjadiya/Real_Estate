# Price Vs Location

# Conclusion:
- Prices are high in Vesu, Hazira, and Dumas Road.
- Prices are moderate in Athwalines, Pal, Althan, etc.
- Prices are low in Kamrej, Katargam, Udhna, etc.
- Outliers can be observed in the prices corresponding to location.

```python
plt.figure(figsize=(12,5))
sns.barplot(x=df['location'], y=df['price'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x=df['location'], y=df['price'])
plt.xticks(rotation='vertical')
plt.show()
```
